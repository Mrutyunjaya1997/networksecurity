```python
def fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        print("In correct output")
    elif n == 0:
        print(0)
    elif n == 1:
        print(1)
    else:
        for i in range(1,n+1):
            c = a + b
            a = b
            b = c
            print(b)
            
        
    
fibonacci(10)
```

```python
import math
def is_prime(n:int):
    if n <= 1:
        print('Not prime')
    else:
        is_prime = True
        for i in range(2,int(math.sqrt(n)) + 1):
            if n % i == 0:
                is_prime = False
                break
        print("Prime") if is_prime else print("Not prime")

is_prime(29)
```

```python
def factorial(n):
    fact = 1
    for i in range(1,n+1):
        fact = fact * i

    return fact

```

```python
print(factorial(5))
```

```python
def rec_factorial(n):
    return 1 if (n == 0 or n == 1) else n * rec_factorial(n-1)

```

```python
print(rec_factorial(3))
```
